```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/S K R I P S I [HASIL]/3. LABELING DATA/DATASET_GDA(labelling).csv')
df
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Memisahkan teks berdasarkan sentimen
positive_text = ' '.join(df[df['Hasil'] == 'Positif']['data_clean'])

# Membuat WordCloud untuk sentimen positif
wordcloud_positive = WordCloud(width=900, height=400, background_color='black', colormap='Blues').generate(positive_text)

# Menampilkan WordCloud
plt.figure(figsize=(20, 10))

plt.subplot(1, 2, 1)
plt.imshow(wordcloud_positive, interpolation='bilinear')
plt.title('Sentimen Positif')
plt.axis('off')

plt.show()
```

```python
# Memisahkan teks berdasarkan sentimen
negative_text = ' '.join(df[df['Hasil'] == 'Negatif']['data_clean'])

# Membuat WordCloud untuk sentimen negatif
wordcloud_negative = WordCloud(width=900, height=400, background_color='black', colormap='Reds').generate(negative_text)

# Menampilkan WordCloud
plt.figure(figsize=(20, 10))

plt.subplot(1, 2, 2)
plt.imshow(wordcloud_negative, interpolation='bilinear')
plt.title('Sentiment Negatif')
plt.axis('off')

plt.show()
```
