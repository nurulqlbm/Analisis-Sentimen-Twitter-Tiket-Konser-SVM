MENGHAPUS DATASET YANG TIDAK DIPERLUKAN

```python
kata1 = 'wtb'
kata2 = 'wts'
kata3 = 'wtt'
kata4 = 'want to buy'
kata5 = 'want to sell'
kata6 = 'want to trade'
kata7 = 'jastip'
kata8 = 'jasa titip'
kata9 = 'testimoni'

df = df_raw[~(df_raw['full_text'].str.contains(kata1, case=False) |
                   df_raw['full_text'].str.contains(kata2, case=False) |
                   df_raw['full_text'].str.contains(kata3, case=False) |
                   df_raw['full_text'].str.contains(kata4, case=False) |
                   df_raw['full_text'].str.contains(kata5, case=False) |
                   df_raw['full_text'].str.contains(kata6, case=False) |
                   df_raw['full_text'].str.contains(kata7, case=False) |
                   df_raw['full_text'].str.contains(kata8, case=False) |
                   df_raw['full_text'].str.contains(kata9, case=False))]

# Hasilnya adalah DataFrame yang tidak mengandung baris dengan kata1, kata2, atau kata3 di kolom 'nama_kolom'
df
```

CLEANSING

```python
import re
import string
```

```python
def Cleansing(text):
    #remove mentions
    text = re.sub(r'@\w+', '', text)
    #remove hashtag
    text = re.sub(r'#\w+', '', text)
    #remove RT
    text = re.sub(r'\bRT\b', '', text)
    #remove link
    text = re.sub(r'http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\\(\\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+', '', text)
    # remove representasi ampersand
    text = re.sub(r'&[a-zA-Z]+;', '', text)
    # remove all characters non alfabet
    text = re.sub(r'[^A-Za-z ]+', '', text)
    # remove alfabet berulang
    text = re.sub(r'(.)\1+', r'\1\1', text)
    #remove emoji
    text = re.sub(r'[^\x00-\x7F]+', '', text)
    # remove new line into space
    text = text.replace('\n', '')
    # remove all punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))
    # remove characers space from both left and right text
    text = text.strip('')

    return text

df['Cleansing'] = df['full_text'].apply(Cleansing)
df
```

CASE FOLDING

```python
def CaseFolding(text):
  text = text.lower()
  return text

df['CaseFolding'] = df['Cleansing'].apply(CaseFolding)
df
```

TOKENIZING

```python
import nltk
nltk.download('punkt')
```

```python
from nltk.tokenize import word_tokenize

def Tokenizing(text):
  text = word_tokenize(text)
  return text

df['Tokenizing'] = df['CaseFolding'].apply(Tokenizing)
df
```

NORMALIZATION

```python
normalisasi_kata= pd.read_csv("/content/drive/MyDrive/S K R I P S I [HASIL]/2. PREPROCESSING/normalisasi.csv")
normalisasi_kata_dict = {}

for index, row in normalisasi_kata.iterrows():
  if row[0] not in normalisasi_kata_dict:
    normalisasi_kata_dict[row[0]] = row[1]

def Normalized(text):
  return [normalisasi_kata_dict[term] if term in normalisasi_kata_dict else term for term in text]

df['Normalized'] = df['Tokenizing'].apply(Normalized)
df
```

STOPWORD

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
stop_words = stopwords.words('indonesian')
```

```python
def Stopword(text):
  return [word for word in text if word not in stop_words]

df['Stopword'] = df['Normalized'].apply(Stopword)
df
```

```python
df['data_clean'] = df['Stopword'].apply(lambda tokens: ' '.join(tokens))
df
```
